```python
#чтение данных, так как названия конфет уникальны, используем их в качестве индекса
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
DATA = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/week3/excercise.csv", delimiter=',')
```

```python
DATA
```

## Task 1


Определите выборочное среднее X, Y


```python
print(np.mean(DATA['X']))
print(np.mean(DATA['Y']))
```

```python
#отбираем данные для предикторов, удаляя два последних столбца, индекс не включается в данные.
X = pd.DataFrame(DATA['X'])
```

```python
#указываем столбец отклика
Y = pd.DataFrame(DATA['Y'])
```

```python
#обучение модели
reg = LinearRegression().fit(X, Y)
```

Значение коэффициента $\theta_0$:

```python
b = reg.intercept_[0]
```

Значение коэффициентов $\theta_1, \ldots, \theta_p$:

```python
a = reg.coef_[0][0]
```

R^2

```python
Y_pred = []
for i in range (10):
  y = a*DATA.iloc[i,1] + b;
  Y_pred.append(y)

r2_score(Y,Y_pred)
```

```python
DATA = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/week3/candy-data.csv", delimiter=',', index_col='competitorname')
DATA
```

```python
#обучение модели будем проводить на данных, за исключением некоторых конфет
train_data = DATA.drop(['One dime','Nik L Nip'])
```

```python
#отбираем данные для предикторов, удаляя два последних столбца, индекс не включается в данные.
X = pd.DataFrame(train_data.drop(['winpercent', 'Y'], axis=1))
X
```

```python
#указываем столбец отклика
y = pd.DataFrame(train_data['winpercent'])
```

```python
#обучение модели
reg = LinearRegression().fit(X, y)
```

```python
#предсказание для конфет введеных вручную
reg.predict([[0, 1, 0, 1, 1, 0, 1, 1, 1, 0.848, 0.594]])
```
